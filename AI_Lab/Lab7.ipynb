GREEDY FIRST SEARCH (GFS)

```python
def h(n):
    H_dist = {
        'A': 11,
        'B': 6,
        'C': 99,
        'D': 1,
        'E': 7,
        'G': 0,
    }
    return H_dist[n]

def GFS(graph,start, goal):
    front = [start]
    visited = set()
    path = []
    while(front):
        #finding node in front with least h(n)
        curr = None
        #finding least h(n) out of nodes in frontier
        for k in front: #for every node in frontier
            if curr== None or h(curr)>h(k): #if there is no current node
                curr = k
        path = path + [curr]
        if curr == goal:
            return path
        front.remove(curr)
        visited.add(curr)
        for adj in graph[curr]:
            if adj not in visited and adj not in front:
                front.append(adj)
    print("Goal node not found")
    return
        
graph = {
    'A': ['B', 'E'],
    'B': ['A', 'C', 'G'],
    'C': ['B'],
    'D': ['E','G'],
    'E': ['A', 'D'],
    'G': ['B', 'D']
}
path = GFS(graph, 'A', 'G')
print("Path to goal node: ", path)
print("program completed")
```
