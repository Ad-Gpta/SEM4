Greedy first search / best first search

```python
def H(k): #heuristic function
    h_val ={
        'A' : 10,
        'B' : 8,
        'C' : 5,
        'D' : 7,
        'E' : 3,
        'F' : 6,
        'G' : 5,
        'H' : 3,
        'I' : 1,
        'J' : 0,
    }
    return h_val[k]

graph = { #no path cost
        'A' : ['B', 'F'],
        'B' : ['A','D','C'],
        'C' : ['B','D','E'],
        'D' : ['B','C','E'],
        'E' : ['I','D','C''J'],
        'F' : ['A','G','H'],
        'G' : ['F','I'],
        'H' : ['F','I'],
        'I' : ['G','H','E','J'],
        'J' : ['E','I'],
    }

q = []
visited = set()

def GFS(start, goal):
    q.append([start, []]) #no path cost
    while(q != []):
        [curr, path] = q[0]
        q.remove(q[0])
        visited.add(curr)
        path = path + [curr]
        if(curr == goal):
            return path
        min_adj = None
        for adj in graph[curr]:
            if adj not in visited:
                if(min_adj == None or H(min_adj)>H(adj)):
                    min_adj = adj  #find neighbour with least heuristic fcn
        q.append([min_adj, path])
    return None

path = GFS('A', 'J')
if path == None:
    print("No path available")
else:
    print("path: ",path)
```

A* algorithm

```python
def H(k): #heuristic function
    h_val ={
        'A' : 10,
        'B' : 8,
        'C' : 5,
        'D' : 7,
        'E' : 3,
        'F' : 6,
        'G' : 5,
        'H' : 3,
        'I' : 1,
        'J' : 0,
    }
    return h_val[k]

graph = {
        'A' : {'B': 6, 'F': 3},
        'B' : {'A':6,'D':2,'C':3},
        'C' : {'B':3,'D':1,'E':5},
        'D' : {'B':2,'C':1,'E':8},
        'E' : {'I':5,'D':8,'C':5,'J':5},
        'F' : {'A':3,'G':1,'H':7},
        'G' : {'F':1,'I':3},
        'H' : {'F':7,'I':2},
        'I' : {'G':3,'H':2,'E':5,'J':3},
        'J' : {'E':5,'I':3},
    }

q = []
visited = set()

def fcn(curr, k):
    h = H(k)
    g = graph[curr][k]
    return h+g

def Astar(start, goal):
    q.append([start, [], 0])
    while(q!= []):
        [curr, path, cost] = q[0] #current node, ancestors, cost to get to node
        q.remove(q[0])
        visited.add(curr)
        path = path + [curr]
        if(curr == goal):
            return path, cost
        min_adj = None
        for adj in graph[curr].keys():
            if adj not in visited:
                if(min_adj == None or fcn(curr, adj)<fcn(curr, min_adj)):
                    min_adj = adj   
        totalcost = cost + graph[curr][min_adj]
        q.append([min_adj, path, totalcost])
    return None, None

path, cost = Astar('A','J')
if path == None:
    print("Goal node not found")
else:
    print("path: ", path,"\ncost: ", cost)
```

Water Jug problem

```python
class Jugs:
    def __init__(self, mj1, mj2, j1, j2):
        self.mj1 =  mj1
        self.mj2 =  mj2
        self.j1 =  j1
        self.j2 =  j2
    def getnext(self):
        next_state = []
        next_state.append(Jugs(self.mj1, self.mj2, self.j1, 0))
        next_state.append(Jugs(self.mj1, self.mj2, 0, self.j2))
        next_state.append(Jugs(self.mj1, self.mj2, self.mj1, self.j2))
        next_state.append(Jugs(self.mj1, self.mj2, self.j1, self.mj2))
        
        s = self.j1+ self.j2
        
        if s <= self.mj1:
            next_state.append(Jugs(self.mj1, self.mj2, s, 0))
        else:
            next_state.append(Jugs(self.mj1, self.mj2, self.mj1, s - self.mj1))
        if s <= self.mj2:
            next_state.append(Jugs(self.mj1, self.mj2, 0, s))
        else:
            next_state.append(Jugs(self.mj1, self.mj2, s - self.mj2, self.mj2))
        print("current state: ", self.j1, self.j2)
        return next_state
    def goaltest(self,goal):
        return self.j1 == goal[0] and self.j2 == goal[1]
    
j = Jugs(4,3,0,0)

def dfs(jug, visited):
    visited.append([jug.j1, jug.j2])
    print("current jug: ", jug.mj1, jug.mj2, jug.j1, jug.j2)
    if(jug.goaltest((2,0))):
        print("final state: ", jug.j1, jug.j2)
        return True
    next_states = jug.getnext()
    for state in next_states:
        if [state.j1, state.j2] not in visited:
            if dfs(state, visited):
                return True
    return False
            
if dfs(j, []):
    print("possible")
else:
    print("not possible")
```
